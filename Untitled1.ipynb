```python
import glob
import pickle
import os
import sys
import numpy as np
from datetime import datetime
```

```python
def chooser(x):
    if x[-6:] == "_final":
        r = x[-21:-6]
    else:
        r = x[-15:]
    
    return datetime.strptime(r, "%Y%m%d-%H%M%S")

        
```

```python
fileList = [f for f in glob.glob(os.path.join(os.path.expanduser('~'),"network_tests",'**'),recursive=True) if not os.path.isdir(f)]
fileList.sort(key=chooser)
saveList = pickle.load(open(fileList[-1],'rb'))
print(fileList[-1])
```

```python
neurons_scatter_x = [f[0][1] for f in saveList]
y = [f[2] for f in saveList]

layers_scatter_x  = [f[0][0] for f in saveList]

filters_scatter_x = [f[0][2] for f in saveList]
```

```python
def duplicates(lst,y, item):
    return np.mean([y[i] for i, x in enumerate(lst) if x == item])

def getMean(x,y):
    r = dict((k, duplicates(x, y, k)) for k in set(x) if x.count(k) > 1)
    list1 = r.keys()
    list2 = r.values()
    keys,values = (list(t) for t in zip(*sorted(zip(list1, list2))))
    return keys,values
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.layouts import column, row
from bokeh.palettes import Spectral11, Category20, Set1
from bokeh.models import Range1d
output_notebook()
```

```python
pn = figure()
pl = figure()
pf = figure()
pn.scatter(neurons_scatter_x,y)
keys,values = getMean(neurons_scatter_x,y)
pn.line(keys,values,line_color = 'red')#, fill_color = "red")
pn.xaxis.axis_label = "Neurons per Layer"
pn.yaxis.axis_label = "Accuracy"

pl.scatter(layers_scatter_x,y)
keys,values = getMean(layers_scatter_x,y)
pl.line(keys,values,line_color = 'red')#, fill_color = "red")
pl.xaxis.axis_label = "Hidden Layers"
pl.yaxis.axis_label = "Accuracy"

pf.scatter(filters_scatter_x,y)
keys,values = getMean(filters_scatter_x,y)
pf.line(keys,values,line_color = 'red')#, fill_color = "red")
pf.xaxis.axis_label = "#Convolution Filters"
pf.yaxis.axis_label = "Accuracy"

show(pn)
show(pl)
show(pf)
```
